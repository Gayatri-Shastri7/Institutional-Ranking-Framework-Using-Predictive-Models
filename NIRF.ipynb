```python
from bs4 import BeautifulSoup
```

```python
import requests as request
```

```python
pageContent = request.get("https://www.nirfindia.org/2021/EngineeringRanking.html")
```

```python
beautifulSoupObject = BeautifulSoup(pageContent.content,'html.parser')
```

```python
print(beautifulSoupObject)
```

```python
print(beautifulSoupObject.prettify())
```

```python
table1 = beautifulSoupObject.find('table', id='tbl_overall')
table1
```

```python
# Obtain every title of columns with tag <th>
headers = []
for i in table1.find_all('td'):   
    title = i.text
    headers.append(title)
```

```python
# print(headers[0],headers[1],headers[2],headers[3],headers[4],headers[5])
print(headers)
```

```python
headers = set(headers)
stopwords = ['More DetailsClose |  | TLR (100)RPC (100)GO (100)OI (100)PERCEPTION (100)']
stopwords = set(stopwords)
final_list = headers - stopwords
final_list = list(final_list)
```

```python
final_list
```

```python
import pandas as pd
```

```python
mydata = pd.DataFrame(columns = headers)
```

```python
for j in table1.find_all('tr')[1:]:
    row_data = j.find_all('th')
    row = [i.text for i in row_data]
    length = len(mydata)
    mydata.loc[length] = row
```
